## 💾 Los datos

La compañia guarda la información que necesita en las siguientes tablas. Algunos de los campos son anónimos para cumplir con los regulaciones de privacidad.

#### Doctors contiene la información sobre los doctores. Cada fila representa un doctor.
- "DoctorID" - es un identificador unico para cada doctor.
- "Region" - la región greográfica actual del doctor.
- "Category" - el tipo de doctor, ya sea 'Specialist' or 'General Practitiones'.
- "Rank" - es un sistema de ranking interno. Es una variable ordenada: El nivel mas alto es Ambassadors, seguido de Titanium Plus, Titanium, Platinum Plus, Platinum, Gold Plus, Gold, Silver Plus, y el nivel más bajo es Silver.
- "Incidence rate" y "R rate" - se relaciona con la cantidad de reprocesamiento que genera cada médico.
- "Satisfaction" - mide la satisfacción que el doctor tiene con la compañia.
- "Experience" - se relaconacon la experiencia del doctor con la compañia.
- "Purchases" - compras en el último año.

#### Orders contiene detalles de las ordenes. Cada fila representa una orden; un doctor puede tener nultiples ordenes.
- "DoctorID" - doctor id (coincide con las otras tablas) (matches the other tables).
- "OrderID" - identificador de orden.
- "OrderNum" - número de orden.
- "Conditions A through J" - mapea las diferentes configuraciones de cada uno de los dispositivos en la orden. Cada orden es para un solo paciente.

#### Complaints collects information on doctor complaints.
#### Complaints recopila información acerca de las quejas de los doctores.
- "DoctorID" - doctor id (coincide con las otras tablas).
- "Complaint Type" - la clasificación que pone la compañia a cada queja.
- "Qty" - número de quejas por tipo de queja por doctor.

#### Instructions has information on whether the doctor includes special instructions on their orders.
#### Instructions tiene información sobre instrucciones especiales en sus pedidos.
- "DoctorID" - doctor id (coincide con las otras tablas). 
- "Instructions" - 'Yes' cuando el doctor incluye instrucciones especiales, 'No' cuand no.
